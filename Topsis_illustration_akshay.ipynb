```python
import matplotlib.pyplot as plt
import pandas as pd

# Load TOPSIS results from CSV file
topsis_results = pd.read_csv("/content/topsis_results.csv")

# Set a custom color palette for better aesthetics
colors = ['#3498db', '#2ecc71']

# Visualize TOPSIS scores
plt.figure(figsize=(12, 6))
plt.bar(topsis_results["Model"], topsis_results["TOPSIS Score"], color=colors[0])
plt.title('TOPSIS Scores for Each Model', fontsize=16, fontweight='bold')
plt.xlabel('Model', fontsize=12)
plt.ylabel('TOPSIS Score', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Visualize TOPSIS ranks
plt.figure(figsize=(12, 6))
plt.bar(topsis_results["Model"], topsis_results["Rank"], color=colors[1])
plt.title('TOPSIS Ranks for Each Model', fontsize=16, fontweight='bold')
plt.xlabel('Model', fontsize=12)
plt.ylabel('Rank', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Visualize comparison of evaluation metrics
metric_names = ["BLEU Score", "Semantic Coherence", "Factual Accuracy", "Content Coverage", "Readability"]

for metric in metric_names:
    plt.figure(figsize=(5, 5))
    plt.bar(topsis_results["Model"], topsis_results[metric], color='crimson')
    plt.title(f'{metric} Across Models')
    plt.xlabel('Model')
    plt.ylabel(metric)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```
